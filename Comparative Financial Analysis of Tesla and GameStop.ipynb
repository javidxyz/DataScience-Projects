The main goals are:

### Extracting Tesla Stock Data Using yFinance

```python
import yfinance as yf
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

### Extracting Tesla Revenue Data Using Web Scraping

```python
import pandas as pd

url ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
tables = pd.read_html(url)
tesla_revenue = tables[1]  
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '']
tesla_revenue.head()
```

### Extracting GameStop Stock Data Using yFinance

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

### Extracting GameStop Revenue Data Using Web Scraping

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
tables = pd.read_html(url)
gme_revenue = tables[1]  # Adjust index if needed
gme_revenue.columns = ['Date', 'Revenue']
gme_revenue = gme_revenue[gme_revenue['Revenue'] != '']
gme_revenue.head()
```

### Tesla Stock and Revenue Dashboard

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=tesla_data['Date'], y=tesla_data['Close'], name="Tesla Stock"))
fig.add_trace(go.Scatter(x=pd.to_datetime(tesla_revenue['Date']), y=tesla_revenue['Revenue'], name="Tesla Revenue"))
fig.update_layout(title='Tesla Stock Price vs Revenue', xaxis_title='Date', yaxis_title='USD')
fig.show()
```

### GameStop Stock and Revenue Dashboard

```python

fig = go.Figure()
fig.add_trace(go.Scatter(x=gme_data['Date'], y=gme_data['Close'], name="GME Stock"))
fig.add_trace(go.Scatter(x=pd.to_datetime(gme_revenue['Date']), y=gme_revenue['Revenue'], name="GME Revenue"))
fig.update_layout(title='GameStop Stock Price vs Revenue', xaxis_title='Date', yaxis_title='USD')
fig.show()
```
